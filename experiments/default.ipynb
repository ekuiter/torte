```python
from IPython.display import Image, display
import pandas as pd
import cv2
import matplotlib.pyplot as plt

output_directory = '../output'

def read_dataframe(stage, file=None):
    if not file:
        file = 'output'
    df = pd.read_csv(f'{output_directory}/{stage}/{file}.csv')
    return df

df = read_dataframe('community-structure')
for file in df['jpg-file'].dropna():
    image = cv2.imread(f'{output_directory}/community-structure/{file}')
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```
