```python
import pandas as pd
import plotly.express as px

file_kmax = '/home/rami/torte/output/kconfig/kmax/output.csv'
file_kconfigreader = '/home/rami/torte/output/kconfig/kconfigreader/output.csv'
file_configfixextractor = '/home/rami/torte/output/kconfig/configfixextractor/output.csv'

def prepare_data(file_path, extractor_name):
    df = pd.read_csv(file_path)
    df['system'] = df['system'].astype(str)
    df['Model-Time (min)'] = df['model-time'] / (60 * 1e9)  
    df['extractor'] = extractor_name

    return df[['system', 'Model-Time (min)', 'extractor']]


df_kmax = prepare_data(file_kmax, 'kmax')
df_kconfigreader = prepare_data(file_kconfigreader, 'kconfigreader')
df_configfixextractor = prepare_data(file_configfixextractor, 'configfixextractor')


df_all = pd.concat([df_kmax, df_kconfigreader, df_configfixextractor])


average_time = df_all.groupby('extractor')['Model-Time (min)'].mean().reset_index()

fig = px.bar(average_time, 
             x='extractor', 
             y='Model-Time (min)', 
             title='Durchschnittliche Verarbeitungszeit für jeden Extractor',
             labels={'extractor': 'Extractor', 'Model-Time (min)': 'Durchschnittliche Verarbeitungszeit (Minuten)'},
             color='extractor',  
             color_discrete_sequence=px.colors.qualitative.Set2)  # Qualitative Farbpalette

fig.update_layout(
    xaxis_title='Extractor',
    yaxis_title='Durchschnittliche Verarbeitungszeit (Minuten)',
    xaxis_tickangle=-45
)

fig.show()


fig = px.bar(df_configfixextractor, 
             x='system',  
             y='Model-Time (min)', 
             title='Verarbeitungszeiten für kmax',
             labels={'system': 'System', 'Model-Time (min)': 'Verarbeitungszeit (Minuten)'},
             color_discrete_sequence=['#40E0D0'])  

fig.update_layout(
    xaxis_title='System',
    yaxis_title='Verarbeitungszeit (Minuten)',
    xaxis_tickangle=-45
)

fig.show()


fig = px.bar(df_kconfigreader, 
             x='system',  
             y='Model-Time (min)', 
             title='Verarbeitungszeiten für kmax',
             labels={'system': 'System', 'Model-Time (min)': 'Verarbeitungszeit (Minuten)'},
             color_discrete_sequence=['#FFBF00']) 

fig.update_layout(
    xaxis_title='System',
    yaxis_title='Verarbeitungszeit (Minuten)',
    xaxis_tickangle=-45
)

fig.show()


fig = px.bar(df_kmax, 
             x='system',  
             y='Model-Time (min)', 
             title='Verarbeitungszeiten für kmax',
             labels={'system': 'System', 'Model-Time (min)': 'Verarbeitungszeit (Minuten)'},
             color_discrete_sequence=['#1f77b4'])  

fig.update_layout(
    xaxis_title='System',
    yaxis_title='Verarbeitungszeit (Minuten)',
    xaxis_tickangle=-45
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Dateien
file_kmax = '/home/rami/torte/output/kconfig/kmax/output.csv'
file_kconfigreader = '/home/rami/torte/output/kconfig/kconfigreader/output.csv'
file_configfixextractor = '/home/rami/torte/output/kconfig/configfixextractor/output.csv'

# Funktion zur Datenaufbereitung
def prepare_data(file_path, extractor_name):
    df = pd.read_csv(file_path)
    df['system'] = df['system'].astype(str)
    df['Model-Time (sec)'] = df['model-time'] / 1e9  # Nanoseconds to seconds
    df['extractor'] = extractor_name
    return df[['system', 'Model-Time (sec)', 'extractor']]

# Daten vorbereiten
df_kmax = prepare_data(file_kmax, 'kmax')
df_kconfigreader = prepare_data(file_kconfigreader, 'kconfigreader')
df_configfixextractor = prepare_data(file_configfixextractor, 'configfixextractor')

# Alle Daten kombinieren
df_all = pd.concat([df_kmax, df_kconfigreader, df_configfixextractor])

# Boxplot mit einer Mittellinie für alle Extractor-Typen
fig = px.box(
    df_all,
    x='extractor',
    y='Model-Time (sec)',
    title='Verteilung der Verarbeitungszeiten nach Extractor (in Sekunden)',
    labels={'extractor': 'Extractor', 'Model-Time (sec)': 'Verarbeitungszeit (Sekunden)'},
    color='extractor',
    color_discrete_sequence=px.colors.qualitative.Set2
)

# Median für die Mittellinie berechnen
median = df_all['Model-Time (sec)'].median()

# Mittellinie hinzufügen
fig.add_shape(
    type="line",
    x0=-0.5,  # Anfang der Linie (leicht außerhalb der Achse)
    x1=2.5,   # Ende der Linie (leicht außerhalb der Achse)
    y0=median,
    y1=median,
    line=dict(color="red", width=2, dash="dash"),
    xref="x",
    yref="y"
)

# Layout anpassen (weißer Hintergrund)
fig.update_layout(
    xaxis_title='Extractor',
    yaxis_title='Verarbeitungszeit (Sekunden)',
    xaxis_tickangle=-45,
    plot_bgcolor='white',  # Hintergrund der Diagrammfläche
    paper_bgcolor='white'  # Hintergrund des gesamten Plots
)

fig.show()

# Boxplot für einzelne Extractor-Typen mit weißem Hintergrund (optional)
for extractor, color in zip(
    ['kmax', 'kconfigreader', 'configfixextractor'], 
    ['#1f77b4', '#FFBF00', '#40E0D0']
):
    subset = df_all[df_all['extractor'] == extractor]
    median = subset['Model-Time (sec)'].median()  # Median für diesen Extractor

    fig = px.box(
        subset,
        x='system',
        y='Model-Time (sec)',
        title=f'Verteilung der Verarbeitungszeiten für {extractor} (in Sekunden)',
        labels={'system': 'System', 'Model-Time (sec)': 'Verarbeitungszeit (Sekunden)'},
        color_discrete_sequence=[color]
    )

    # Mittellinie hinzufügen
    fig.add_shape(
        type="line",
        x0=-0.5,
        x1=len(subset['system'].unique()) - 0.5,
        y0=median,
        y1=median,
        line=dict(color="red", width=2, dash="dash"),
        xref="x",
        yref="y"
    )

    # Layout anpassen (weißer Hintergrund)
    fig.update_layout(
        xaxis_title='System',
        yaxis_title='Verarbeitungszeit (Sekunden)',
        xaxis_tickangle=-45,
        plot_bgcolor='white',  # Hintergrund der Diagrammfläche
        paper_bgcolor='white'  # Hintergrund des gesamten Plots
    )

    fig.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Datei einlesen
file_configfixextractor = '/home/rami/torte/output/kconfig/output.csv'
df = pd.read_csv(file_configfixextractor)

# Konvertiere die Zeit in numerische Werte
df['model-time'] = pd.to_numeric(df['model-time'], errors='coerce')

# Konvertiere die 'model-time' von Millisekunden in Minuten
df['model-time-minutes'] = df['model-time'] / 1e9   # Zeit in Minuten umrechnen

# Status basierend auf den vorhandenen Werten für 'model-file' und 'model-time'
df['status'] = df.apply(
    lambda row: 'Extracted' if pd.notna(row['model-file']) and pd.notna(row['model-time']) else 'Not Extracted',
    axis=1
)

# Zusammenfassung der Anzahl der 'status' pro 'system'
df_summary = df.groupby(['system', 'status']).size().reset_index(name='count')

# Balkendiagramm
fig_bar = px.bar(
    df_summary,
    x='system',
    y='count',
    color='status',
    title='Machbarkeit der Feature-Modell-Extraktion pro System',
    labels={'count': 'Anzahl', 'system': 'System', 'status': 'Status'},
    text='count',  
)

fig_bar.update_layout(
    xaxis_title='System',
    yaxis_title='Anzahl der Revisionen',
    legend_title='Extraction Status',
    xaxis_tickangle=-45,
    barmode='stack',  
    plot_bgcolor='white',  
    paper_bgcolor='white',  
)


fig_bar.show()

# Boxplot für die Verteilung der Zeiten in Minuten
fig_box = px.box(
    df,
    x='system',
    y='model-time-minutes',
    color='status',
    title='Verteilung der Model-Zeiten pro System (in Minuten)',
    labels={'model-time-minutes': 'Model Zeit (in Minuten)', 'system': 'System', 'status': 'Status'}
)

# Berechnung der Mittellinie (Median)
median = df['model-time-minutes'].median()

# Mittellinie hinzufügen
fig_box.add_shape(
    type="line",
    x0=-0.5,
    x1=len(df['system'].unique()) - 0.5,
    y0=median,
    y1=median,
    line=dict(color="red", width=2, dash="dash"),
    xref="x",
    yref="y"
)

fig_box.update_layout(
    xaxis_title='System',
    yaxis_title='Model Zeit (in Minuten)',
    legend_title='Extraction Status',
    plot_bgcolor='white',  
    paper_bgcolor='white',  
)

# Anzeigen beider Diagramme
fig_box.show()
```

```python
import pandas as pd
import plotly.express as px

# Datei einlesen
file_configfixextractor = '/home/rami/torte/output/kconfig/output.csv'
df = pd.read_csv(file_configfixextractor)

# Konvertiere die Zeit in numerische Werte
df['model-time'] = pd.to_numeric(df['model-time'], errors='coerce')

# Konvertiere die 'model-time' von Millisekunden in Minuten
df['model-time-minutes'] = df['model-time'] / 60000  # Zeit in Minuten umrechnen

# Status basierend auf den vorhandenen Werten für 'model-file' und 'model-time'
df['status'] = df.apply(
    lambda row: 'extrahierte Feature-Modelle' if pd.notna(row['model-file']) and pd.notna(row['model-time']) else 'Nicht extrahierte Feature-Modelle',
    axis=1
)

# Zusammenfassung der Anzahl der 'status' pro 'system'
df_summary = df.groupby(['system', 'status']).size().reset_index(name='count')

# Balkendiagramm
fig_bar = px.bar(
    df_summary,
    x='system',
    y='count',
    color='status',
    title='Machbarkeit der Feature-Modell-Extraktion pro System',
    labels={'count': '', 'system': 'System', 'status': 'Status'},
    text='count',  
)

fig_bar.update_layout(
    xaxis_title='System',
    legend_title='Extraction Status',
    xaxis_tickangle=-45,
    barmode='stack',  
    plot_bgcolor='white',  
    paper_bgcolor='white',
    yaxis=dict(
        showticklabels=False,  
        showgrid=False         
    )
)

fig_bar.show()
```

```python
import pandas as pd
import plotly.express as px

# Datei einlesen
file_configfixextractor = '/home/rami/torte/output/kconfig/output.csv'
df = pd.read_csv(file_configfixextractor)

# Konvertiere die Zeit in numerische Werte und in Sekunden
df['model-time'] = pd.to_numeric(df['model-time'], errors='coerce')
df['model-time-seconds'] = df['model-time'] / 1e9  # Zeit von Nanosekunden in Sekunden umrechnen

# Status basierend auf den vorhandenen Werten für 'model-file' und 'model-time'
df['status'] = df.apply(
    lambda row: 'Extracted' if pd.notna(row['model-file']) and pd.notna(row['model-time']) else 'Not Extracted',
    axis=1
)

# Median für die Mittellinie berechnen
median = df['model-time-seconds'].median()

# Boxplot mit Plotly Express
fig = px.box(
    df,
    x='system',  # Systeme auf der X-Achse
    y='model-time-seconds',  # Zeit in Sekunden auf der Y-Achse
    color='status',  # Farben für Status ('Extracted' vs. 'Not Extracted')
    title='Algorithm Runtime (in seconds, log scale)',
    labels={
        'system': 'System',
        'model-time-seconds': 'Runtime (seconds)',
        'status': 'Extraction Status'
    },
    notched=True  # Kerben im Boxplot aktivieren
)

# Mittellinie hinzufügen
fig.add_shape(
    type="line",
    x0=-0.5,  # Anfang der Linie (leicht außerhalb der Achse)
    x1=len(df['system'].unique()) - 0.5,  # Dynamisches Ende basierend auf Anzahl Systeme
    y0=median,
    y1=median,
    line=dict(color="red", width=2, dash="dash"),
    xref="x",
    yref="y"
)

# Layout anpassen - Logarithmische Y-Achse mit definierten Ticks
fig.update_layout(

    xaxis=dict(
        title='System',  # Titel für die X-Achse
        tickangle=45  # Leichte Neigung der X-Achsen-Beschriftungen
    ),
    plot_bgcolor='white',  # Weißer Hintergrund für das Diagramm
    paper_bgcolor='white',  # Weißer Hintergrund für die gesamte Fläche
    legend=dict(
        title='Status',
        orientation='h',  # Horizontale Anordnung der Legende
        x=0.5,  # Zentrierte Position der Legende
        xanchor='center',
        y=-0.2
    )
)

# Diagramm anzeigen
fig.show()
```
